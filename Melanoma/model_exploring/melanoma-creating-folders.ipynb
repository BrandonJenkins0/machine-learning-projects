```python
# Importing modules
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os, shutil
```

```python
# Creating directories
train_meta = pd.read_csv("../input/siim-isic-melanoma-classification/train.csv")
train_mal_img_names = train_meta.loc[train_meta["target"] == 1, "image_name"].values + ".jpg"
train_ben_img_names = train_meta.loc[train_meta["target"] == 0, "image_name"].values + ".jpg"
```

```python
# Grabbing 10% of malignant and benign images in train set to create validation set
split_train_mal = np.floor(train_mal_img_names.shape[0] * .1).astype('int')
split_train_ben = np.floor(train_ben_img_names.shape[0] * .1).astype('int')
```

```python
# Trying to create directories baby!!
original_dir = "../input/siim-isic-melanoma-classification/jpeg"

base_dir = "melanoma"
os.mkdir(base_dir)

train_dir = os.path.join(base_dir, "train")
os.mkdir(train_dir)
val_dir = os.path.join(base_dir, "val")
os.mkdir(val_dir)

train_mal_dir = os.path.join(train_dir, 'malignant')
os.mkdir(train_mal_dir)

train_ben_dir = os.path.join(train_dir, 'benign')
os.mkdir(train_ben_dir)

val_mal_dir = os.path.join(val_dir, 'malignant')
os.mkdir(val_mal_dir)

val_ben_dir = os.path.join(val_dir, 'benign')
os.mkdir(val_ben_dir)
```

```python
# Copying images over to dirs
for fname in train_mal_img_names[split_train_mal:]:
    src = os.path.join(original_dir, "train/" + fname)
    dst = os.path.join(train_mal_dir, fname)
    shutil.copyfile(src, dst)

for fname in train_ben_img_names[split_train_ben:]:
    src = os.path.join(original_dir, "train/" + fname)
    dst = os.path.join(train_ben_dir, fname)
    shutil.copyfile(src, dst)

for fname in train_mal_img_names[:split_train_mal]:
    src = os.path.join(original_dir, "train/" + fname)
    dst = os.path.join(val_mal_dir, fname)
    shutil.copyfile(src, dst)

for fname in train_ben_img_names[:split_train_ben]:
    src = os.path.join(original_dir, "train/" + fname)
    dst = os.path.join(val_ben_dir, fname)
    shutil.copyfile(src, dst)
```

```python
# Copying over the test folder
src = os.path.join(original_dir, "test")
dst = os.path.join(base_dir, "test")
shutil.copyfile(src, dst)
```
